```python
# get P(k) at redhsift z=0
import numpy as np
kk = np.logspace(-4,np.log10(3),1000) # k in h/Mpc
kk1 = np.logspace(-4,np.log10(3),1000) # k in h/Mpc
kk2 = np.logspace(-4,np.log10(3),1000) # k in h/Mpc
kk3 = np.logspace(-4,np.log10(3),1000) # k in h/Mpc
kk4 = np.logspace(-4,np.log10(3),1000) # k in h/Mpc

Pk = [] # P(k) in (Mpc/h)**3
h = LambdaCDM.h() # get reduced Hubble for conversions to 1/Mpc
for k in kk:
    Pk.append(LambdaCDM.pk(k*h,0.)*h**3) # function .pk(k,z)

Pk1 = [] # P(k) in (Mpc/h)**3
h1 = LambdaCDM1.h() # get reduced Hubble for conversions to 1/Mpc
for k in kk1:
    Pk1.append(LambdaCDM1.pk(k*h1,0.)*h1**3) # function .pk(k,z)

Pk2 = [] # P(k) in (Mpc/h)**3
h2 = LambdaCDM2.h() # get reduced Hubble for conversions to 1/Mpc
for k in kk2:
    Pk2.append(LambdaCDM2.pk(k*h2,0.)*h2**3) # function .pk(k,z)

Pk3 = [] # P(k) in (Mpc/h)**3
h3 = LambdaCDM3.h() # get reduced Hubble for conversions to 1/Mpc
for k in kk3:
    Pk3.append(LambdaCDM3.pk(k*h3,0.)*h3**3) # function .pk(k,z)

Pk4 = [] # P(k) in (Mpc/h)**3
h4 = LambdaCDM4.h() # get reduced Hubble for conversions to 1/Mpc
for k in kk4:
    Pk4.append(LambdaCDM4.pk(k*h4,0.)*h4**3) # function .pk(k,z)
```

```python
# plot P(k)
plt.figure(2)
plt.xscale('log');plt.yscale('log');plt.xlim(kk[0],kk[-1])
plt.xlabel(r'$k \,\,\,\, [h/\mathrm{Mpc}]$')
plt.ylabel(r'$P(k) \,\,\,\, [\mathrm{Mpc}/h]^3$')

plt.plot(kk,Pk,label=r'$\Lambda$CDM')
plt.plot(kk1,Pk1,label=r'$r=0.01$')
plt.plot(kk2,Pk2,label=r'$r=0.02$')
plt.plot(kk3,Pk3,label=r'$r=0.03$')
plt.plot(kk4,Pk4,label=r'$r=0.04$')

plt.title(r'$P(k)$ with varying $r$')
plt.grid()
plt.legend(loc='upper right')
plt.savefig('pk_r.png',bbox_inches='tight')
```
