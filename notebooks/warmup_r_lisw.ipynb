```python
# uncomment to get plots displayed in notebook
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
from math import pi
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
rc('text', usetex=True)

font = {'size'   : 16, 'family':'Times New Roman'}
axislabelfontsize='large'
matplotlib.rc('font', **font)
matplotlib.mathtext.rcParams['legend.fontsize']='medium'
plt.rcParams["figure.figsize"] = [8.0,6.0]
```

```python
# plot C_l^TT
plt.figure(1)
plt.xscale('log');plt.yscale('linear');plt.xlim(2,2500)
plt.xlabel(r'$\ell$')
plt.ylabel(r'$10^{10} [\ell(\ell+1)/2\pi]  C_\ell^\mathrm{TT}$')
plt.plot(ll,10**10*clTT*ll*(ll+1)/2./pi,label=r'$\Lambda$CDM')
plt.plot(ll1,10**10*clTT1*ll1*(ll1+1)/2./pi,label=r'$r=0.01$')
plt.plot(ll2,10**10*clTT2*ll2*(ll2+1)/2./pi,label=r'$r=0.02$')
plt.plot(ll3,10**10*clTT3*ll3*(ll3+1)/2./pi,label=r'$r=0.03$')
plt.plot(ll4,10**10*clTT4*ll4*(ll4+1)/2./pi,label=r'$r=0.04$')

plt.title(r'late-time ISW effect part of $C_\ell^\mathrm{TT}$ with varying $r$')
plt.grid()
plt.legend(loc='upper right')
plt.savefig('cltt_lisw.png',bbox_inches='tight')
```
